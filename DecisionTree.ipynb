```python
import pandas as pd
movieInfo=pd.read_csv('/content/Movie Interests.csv')
movieInfo
```

```python
inputDataset=movieInfo.drop(columns=['Interest'])
```

```python
inputDataset
```

```python
outputDataSet=movieInfo['Interest']
outputDataSet
```

```python
from sklearn.tree import DecisionTreeClassifier
movieModel=DecisionTreeClassifier()
movieModel.fit(inputDataset,outputDataSet)
movieInterest=movieModel.predict([[9,1],[33,0]])
movieInterest
```
